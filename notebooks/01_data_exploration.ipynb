DATA PRE-PROCESSING

```python
import pandas as pd
file_path = '../data/raw/kidney_disease.csv'
df = pd.read_csv(file_path)
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
df['pcv'] = pd.to_numeric(df['pcv'], errors='coerce')
df['wc'] = pd.to_numeric(df['wc'], errors='coerce')
df['rc'] = pd.to_numeric(df['rc'], errors='coerce')

df.isnull().sum()
```

```python
#separating columns into numbers and catagories
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object', 'str']).columns

#filling numerical missing cells with medians
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

#filling categorical missing cells with modes
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df.isnull().sum().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = df. select_dtypes(include=['object', 'str']).columns

for cols in cat_cols:
    df[cols] = le.fit_transform(df[col])

df.head()
```

```python
df.to_csv('../data/processed/cleaned_kidney_data.csv', index=False)

print("Sample purified and stored in data/processed!")
```
